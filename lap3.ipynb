```python
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

```

```python

texts = [
    "I love learning AI!",
    "This is the worst movie ever.",
    "Hugging Face makes NLP easy."
]

inputs = tokenizer(
    texts,
    padding=True,          # căn chỉnh độ dài tensor
    truncation=True,       # cắt nếu câu quá dài
    return_tensors="pt"    # trả về PyTorch tensor
)

with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits
    probs = torch.softmax(logits, dim=1)
```

```python
labels = model.config.id2label

for text, prob in zip(texts, probs):
    label_id = torch.argmax(prob).item()
    print(text, "->", labels[label_id], prob[label_id].item())
```
